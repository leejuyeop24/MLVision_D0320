```python
## [1-1] 모듈 로딩
import cv2							# 컴퓨터버전 모듈
import numpy as np					# 이미지 데이터 저장 모듈
import matplotlib.pyplot as plt		# 시각화 모듈
import os							# 파일, 폴더, 경로 관련 모듈
```

```python
## [1-2] 데이터 준비
IMG_DIR = '../data/image/'
FILENAME = 'orange.png'
```

[2] 이미지 로딩 <hr>

```python
## 로딩 : 경로+파일명
colorIMG=cv2.imread(IMG_DIR+FILENAME, cv2.IMREAD_COLOR)
```

```python
## 정보확인
if colorIMG is None:
	print(f"{colorIMG}: 이미지 체크")
else:
	print(f'{FILENAME}')
	print(f'형태:{colorIMG.shape}	차원:{colorIMG.ndim}D')

## 이미지 형태 : (행, 열, 채널)
```

```python
## 이미지 확인
rgbIMG = cv2.cvtColor(colorIMG, cv2.COLOR_BGR2RGB)	# BGR 채널 순서 => RGB 채널 순서

plt.imshow(rgbIMG)
plt.grid()
plt.show()
```

[3] 이미지 자르기 <hr>

```python
## 이미지에서 자를 크기 지정

## - 현재 이미지 크기
H, W, _ = colorIMG.shape
print(f'현재 범위 : { H }, {W}')

## - 자르고 싶은 크기
X, Y, nW, nH = 180, 20, 340, 360

## - 가능한 범위 여부 체크
isValid = (X+nW <= W) and (Y+nH <= H)
print(f'가능한 범위 : { isValid }')
```

```python
## 원하는 부분 즉, ROI 추출
## X, Y, nW, nH = 180, 20, 340, 360
roiIMG = colorIMG[Y:Y+nH, X:X+nW]	# [180:520, 20:380]

print(roiIMG.shape)
```

```python
## - 시각화 확인
plt.imshow(	cv2.cvtColor(colorIMG, cv2.COLOR_BGR2RGB) )
plt.show()
```

```python
## - 시각화 확인
plt.imshow(	cv2.cvtColor(roiIMG, cv2.COLOR_BGR2RGB) )
plt.show()
```
